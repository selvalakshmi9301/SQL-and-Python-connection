```python
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
driver=webdriver.Firefox()
driver.get('https://www.flipkart.com/air-conditioners/pr?sid=j9e,abm,c54&p[]=facets.fulfilled_by%255B%255D%3DFlipkart%2BAssured&p[]=facets.brand%255B%255D%3DLG&p[]=facets.serviceability%5B%5D%3Dtrue&otracker=categorytree&otracker=nmenu_sub_TVs%20%26%20Appliances_0_LG&otracker=nmenu_sub_TVs%20%26%20Appliances_0_LG')
```

```python
driver
```

```python
html_content=driver.page_source
html_content
```

```python
Soup=BeautifulSoup(html_content,'html.parser')
Soup
```

```python
AC=Soup.find_all('div',class_='_75nlfW')
AC
```

```python
print(len(AC))
```

```python
brand_Name=[]
rating_Star=[]
ratings=[]
reviews=[]
price=[]
description=[]
for i in AC:
     a=i.find('div',class_='KzDlHZ')
     brand_Name.append(a.string)
     b=i.find('div',class_='XQDdHH')
     rating_Star.append(b.text)
     c=i.find('span',class_='Wphh3N')
     d=c.text
     rating,review=d.split('\xa0&\xa0')
     ratings.append(rating.replace('Ratings',''))
     reviews.append(review.replace('Reviews',''))
     e=i.find('div',class_='Nx9bqj _4b5DiR')
     price.append(e.string)
     f=i.find('div',class_='_6NESgJ')
     description.append(f.text)
print(brand_Name) 
print(rating_Star)
print(ratings)
print(reviews)
print(price)
print(description)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({'Brand_Name':brand_Name,'Rating_Star':rating_Star,'Ratings':ratings,'Reviews':reviews,'Price':price,'Description':description})
df
```
